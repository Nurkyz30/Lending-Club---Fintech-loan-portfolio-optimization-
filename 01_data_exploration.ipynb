```python
# –°–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –∫–æ–ª–æ–Ω–æ–∫
print(f"–í—Å–µ–≥–æ –∫–æ–ª–æ–Ω–æ–∫: {len(accepted.columns)}\n")
print("–°–ø–∏—Å–æ–∫ –∫–æ–ª–æ–Ω–æ–∫:")
for i, col in enumerate(accepted.columns, 1):
    print(f"{i}. {col}")
```

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
print("–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (—Ç–æ–ø-20 –∫–æ–ª–æ–Ω–æ–∫):")
missing = accepted.isnull().sum()
missing_percent = (missing / len(accepted)) * 100
missing_df = pd.DataFrame({
    '–ö–æ–ª–æ–Ω–∫–∞': missing.index,
    '–ü—Ä–æ–ø—É—Å–∫–æ–≤': missing.values,
    '–ü—Ä–æ—Ü–µ–Ω—Ç': missing_percent.values
})
missing_df = missing_df[missing_df['–ü—Ä–æ–ø—É—Å–∫–æ–≤'] > 0].sort_values('–ü—Ä–æ–ø—É—Å–∫–æ–≤', ascending=False)
print(missing_df.head(20))
```

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –∫–ª—é—á–µ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫ 
key_columns = [
    'loan_amnt',           # –°—É–º–º–∞ –∫—Ä–µ–¥–∏—Ç–∞
    'funded_amnt',         # –ü—Ä–æ—Ñ–∏–Ω–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–∞—è —Å—É–º–º–∞
    'int_rate',            # –ü—Ä–æ—Ü–µ–Ω—Ç–Ω–∞—è —Å—Ç–∞–≤–∫–∞
    'installment',         # –ï–∂–µ–º–µ—Å—è—á–Ω—ã–π –ø–ª–∞—Ç—ë–∂
    'grade',               # –ö—Ä–µ–¥–∏—Ç–Ω—ã–π —Ä–µ–π—Ç–∏–Ω–≥ (A-G)
    'sub_grade',           # –ü–æ–¥—Ä–µ–π—Ç–∏–Ω–≥
    'emp_length',          # –°—Ç–∞–∂ —Ä–∞–±–æ—Ç—ã
    'home_ownership',      # –í–ª–∞–¥–µ–Ω–∏–µ –∂–∏–ª—å—ë–º
    'annual_inc',          # –ì–æ–¥–æ–≤–æ–π –¥–æ—Ö–æ–¥
    'loan_status',         # –°—Ç–∞—Ç—É—Å –∫—Ä–µ–¥–∏—Ç–∞ (–ö–õ–Æ–ß–ï–í–ê–Ø!)
    'purpose',             # –¶–µ–ª—å –∫—Ä–µ–¥–∏—Ç–∞
    'addr_state',          # –®—Ç–∞—Ç
    'dti',                 # Debt-to-income ratio
    'delinq_2yrs',         # –ü—Ä–æ—Å—Ä–æ—á–∫–∏ –∑–∞ 2 –≥–æ–¥–∞
    'issue_d',             # –î–∞—Ç–∞ –≤—ã–¥–∞—á–∏
    'term'                 # –°—Ä–æ–∫ –∫—Ä–µ–¥–∏—Ç–∞
]

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è –∫–ª—é—á–µ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫:")
for col in key_columns:
    if col in accepted.columns:
        print(f"‚úÖ {col}")
    else:
        print(f"‚ùå {col} - –û–¢–°–£–¢–°–¢–í–£–ï–¢")
```

```python
# –°–º–æ—Ç—Ä–∏–º —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è loan_status
print("–°—Ç–∞—Ç—É—Å—ã –∫—Ä–µ–¥–∏—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ:")
print(accepted['loan_status'].value_counts())
print("\n")
print("–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:")
print(accepted['loan_status'].value_counts(normalize=True) * 100)
```

```python
# –û—Å–Ω–æ–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –∫–ª—é—á–µ–≤—ã–º –∫–æ–ª–æ–Ω–∫–∞–º
key_numeric = ['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 
               'annual_inc', 'dti', 'fico_range_low', 'fico_range_high']

print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ø–æ –∫–ª—é—á–µ–≤—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º:")
accepted[key_numeric].describe()
```

```python
# –û–ø—Ä–µ–¥–µ–ª—è–µ–º "–ø–ª–æ—Ö–∏–µ" –∫—Ä–µ–¥–∏—Ç—ã (–¥–µ—Ñ–æ–ª—Ç)
bad_statuses = ['Charged Off', 'Default', 
                'Does not meet the credit policy. Status:Charged Off']

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º "—Ö–æ—Ä–æ—à–∏–µ" –∫—Ä–µ–¥–∏—Ç—ã (–≤—ã–ø–ª–∞—á–µ–Ω—ã)
good_statuses = ['Fully Paid', 
                 'Does not meet the credit policy. Status:Fully Paid']

# –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—É—é –∫–æ–ª–æ–Ω–∫—É loan_outcome
def categorize_loan(status):
    if status in bad_statuses:
        return 'Bad'
    elif status in good_statuses:
        return 'Good'
    else:
        return 'Unknown'  # Current, Late, In Grace Period

accepted['loan_outcome'] = accepted['loan_status'].apply(categorize_loan)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ loan_outcome:")
print(accepted['loan_outcome'].value_counts())
print("\n")
print("–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:")
print(accepted['loan_outcome'].value_counts(normalize=True).round(4) * 100)

# Default Rate (—Å—Ä–µ–¥–∏ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—ã—Ö –∫—Ä–µ–¥–∏—Ç–æ–≤)
completed = accepted[accepted['loan_outcome'] != 'Unknown']
default_rate = (completed['loan_outcome'] == 'Bad').sum() / len(completed) * 100
print(f"\nüìä Default Rate (—Å—Ä–µ–¥–∏ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—ã—Ö –∫—Ä–µ–¥–∏—Ç–æ–≤): {default_rate:.2f}%")
```

```python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –≤–∞–∂–Ω—ã–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
print("1. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ grade (–∫—Ä–µ–¥–∏—Ç–Ω—ã–π —Ä–µ–π—Ç–∏–Ω–≥):")
print(accepted['grade'].value_counts().sort_index())
print("\n" + "="*50 + "\n")

print("2. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ term (—Å—Ä–æ–∫ –∫—Ä–µ–¥–∏—Ç–∞):")
print(accepted['term'].value_counts())
print("\n" + "="*50 + "\n")

print("3. –¢–æ–ø-10 —Ü–µ–ª–µ–π –∫—Ä–µ–¥–∏—Ç–∞:")
print(accepted['purpose'].value_counts().head(10))
```

```python
# –°–º–æ—Ç—Ä–∏–º –ø–µ—Ä–∏–æ–¥ –¥–∞–Ω–Ω—ã—Ö
print("–ü–µ—Ä–∏–æ–¥ –¥–∞–Ω–Ω—ã—Ö:")

# –°–Ω–∞—á–∞–ª–∞ –ø—Ä–æ–≤–µ—Ä–∏–º, –µ—Å—Ç—å –ª–∏ –ø—Ä–æ–ø—É—Å–∫–∏
print(f"–ü—Ä–æ–ø—É—Å–∫–æ–≤ –≤ issue_d: {accepted['issue_d'].isnull().sum()}")
print(f"–ü–µ—Ä–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è issue_d:")
print(accepted['issue_d'].head(10))
print("\n")

# –ö–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –≤ datetime, –∏–≥–Ω–æ—Ä–∏—Ä—É—è –æ—à–∏–±–∫–∏
accepted['issue_date'] = pd.to_datetime(accepted['issue_d'], errors='coerce')

# –£–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ –¥–∞—Ç–µ
valid_dates = accepted[accepted['issue_date'].notna()]

print(f"–ü–µ—Ä–≤–∞—è –≤—ã–¥–∞—á–∞: {valid_dates['issue_date'].min()}")
print(f"–ü–æ—Å–ª–µ–¥–Ω—è—è –≤—ã–¥–∞—á–∞: {valid_dates['issue_date'].max()}")

# –°–æ–∑–¥–∞–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –≥–æ–¥–æ–º
accepted['issue_year'] = accepted['issue_date'].dt.year

print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π –ø–æ –≥–æ–¥–∞–º:")
print(accepted['issue_year'].value_counts().sort_index())
```

```python
# –°–æ–∑–¥–∞–µ–º –æ—á–∏—â–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
# –û—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—ã–µ –∫—Ä–µ–¥–∏—Ç—ã (Good/Bad)
df_completed = accepted[accepted['loan_outcome'].isin(['Good', 'Bad'])].copy()

print(f"–î–∞—Ç–∞—Å–µ—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞:")
print(f"–í—Å–µ–≥–æ –∑–∞–ø–∏—Å–µ–π: {len(df_completed):,}")
print(f"–ü–µ—Ä–∏–æ–¥: 2007-2018")
print(f"Default Rate: {(df_completed['loan_outcome'] == 'Bad').sum() / len(df_completed) * 100:.2f}%")
print("\n")

# –í—ã–±–∏—Ä–∞–µ–º –∫–ª—é—á–µ–≤—ã–µ –∫–æ–ª–æ–Ω–∫–∏ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
key_features = [
    # –ò–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä—ã
    'id',
    
    # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫—Ä–µ–¥–∏—Ç–µ
    'loan_amnt', 'funded_amnt', 'term', 'int_rate', 'installment', 'grade', 'sub_grade',
    
    # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∑–∞–µ–º—â–∏–∫–µ
    'emp_length', 'home_ownership', 'annual_inc', 'verification_status',
    
    # –ö—Ä–µ–¥–∏—Ç–Ω–∞—è –∏—Å—Ç–æ—Ä–∏—è
    'dti', 'delinq_2yrs', 'fico_range_low', 'fico_range_high',
    'inq_last_6mths', 'open_acc', 'pub_rec', 'revol_bal', 'revol_util', 'total_acc',
    
    # –¶–µ–ª—å –∏ –≥–µ–æ–≥—Ä–∞—Ñ–∏—è
    'purpose', 'addr_state', 'zip_code',
    
    # –í—Ä–µ–º–µ–Ω–Ω—ã–µ –º–µ—Ç–∫–∏
    'issue_d', 'issue_date', 'issue_year',
    
    # –¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è
    'loan_status', 'loan_outcome'
]

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç
df_analysis = df_completed[key_features].copy()

print(f"–ö–æ–ª–æ–Ω–æ–∫ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞: {len(key_features)}")
print(f"\n–†–∞–∑–º–µ—Ä —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞: {df_analysis.shape}")
print(f"–ü–∞–º—è—Ç—å: {df_analysis.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç
df_analysis.to_csv('lending_club_cleaned.csv', index=False)
print("\n‚úÖ –î–∞–Ω–Ω—ã–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ 'lending_club_cleaned.csv'")
```

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –∫–ª—é—á–µ–≤—ã—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö
print("–ü—Ä–æ–ø—É—Å–∫–∏ –≤ –∫–ª—é—á–µ–≤—ã—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö:")
missing = df_analysis.isnull().sum()
missing_pct = (missing / len(df_analysis) * 100).round(2)
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing': missing.values,
    'Percent': missing_pct.values
})
missing_df = missing_df[missing_df['Missing'] > 0].sort_values('Missing', ascending=False)
print(missing_df)
```

```python
# –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö
critical_cols = ['loan_amnt', 'int_rate', 'grade', 'dti', 'annual_inc', 'loan_outcome']

print(f"–î–æ –æ—á–∏—Å—Ç–∫–∏: {len(df_analysis):,} —Å—Ç—Ä–æ–∫")

# –£–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ–∫–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö
df_clean = df_analysis.dropna(subset=critical_cols).copy()

print(f"–ü–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –ø—Ä–æ–ø—É—Å–∫–æ–≤: {len(df_clean):,} —Å—Ç—Ä–æ–∫")
print(f"–£–¥–∞–ª–µ–Ω–æ: {len(df_analysis) - len(df_clean):,} —Å—Ç—Ä–æ–∫")

# –î–ª—è –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –∫–æ–ª–æ–Ω–æ–∫ –∑–∞–ø–æ–ª–Ω—è–µ–º
df_clean['emp_length'] = df_clean['emp_length'].fillna('Unknown')
df_clean['revol_util'] = df_clean['revol_util'].fillna(df_clean['revol_util'].median())
df_clean['inq_last_6mths'] = df_clean['inq_last_6mths'].fillna(0)
df_clean['open_acc'] = df_clean['open_acc'].fillna(df_clean['open_acc'].median())
df_clean['delinq_2yrs'] = df_clean['delinq_2yrs'].fillna(0)
df_clean['total_acc'] = df_clean['total_acc'].fillna(df_clean['total_acc'].median())
df_clean['pub_rec'] = df_clean['pub_rec'].fillna(0)
df_clean['zip_code'] = df_clean['zip_code'].fillna('Unknown')

print(f"\n‚úÖ –î–∞–Ω–Ω—ã–µ –æ—á–∏—â–µ–Ω—ã!")
print(f"–§–∏–Ω–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä: {df_clean.shape}")

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å –≤ –∫–ª—é—á–µ–≤—ã—Ö –∫–æ–ª–æ–Ω–∫–∞—Ö
print(f"\n–ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ—Å–ª–µ –æ—á–∏—Å—Ç–∫–∏:")
print(df_clean[critical_cols].isnull().sum())

# –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –æ—á–∏—â–µ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç
df_clean.to_csv('lending_club_final.csv', index=False)
print("\n‚úÖ –§–∏–Ω–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ 'lending_club_final.csv'")
```
